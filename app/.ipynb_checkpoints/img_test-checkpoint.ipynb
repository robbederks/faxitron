```python
from matplotlib import pyplot as plt
%matplotlib notebook

import numpy as np
import math
```

```python
dat = []
with open("resp.dat", 'r') as f:
  for l in f:
    dat.append(int(l))

frame = []
for i in range(370, (len(dat)//2)*2, 2):
    frame.append((dat[i+1] << 8 | dat[i]) & 0b111111111111)

plt.figure()
plt.plot(frame)
plt.show()

COLS = 1026
rows = len(frame) // COLS
print(rows)
raw_img = np.reshape(frame[:COLS*rows], (COLS, rows))

plt.figure()
plt.imshow(raw_img, cmap="Greys")
plt.show()
```

```python
# Histogram
#plt.hist(frame, bins=list(range(0, 2 << 9)))

# Take out clipped
usable_frame = list(filter(lambda v: v < 4090, frame))

plt.figure()
plt.hist(usable_frame, bins=4095)
plt.show()
```

```python
import math

def image_histogram_equalization(image, number_bins=4095):
    # from http://www.janeriksolem.net/histogram-equalization-with-python-and.html
    
    # get image histogram
    image_histogram, bins = np.histogram(list(filter(lambda v: v < 4090, image.flatten())), number_bins, density=True)
    cdf = image_histogram.cumsum() # cumulative distribution function
    cdf = 255 * cdf / cdf[-1] # normalize

    # use linear interpolation of cdf to find new pixel values
    image_equalized = np.interp(image.flatten(), bins[:-1], cdf)

    return image_equalized.reshape(image.shape), cdf

def image_histogram_equalization_ceiling(image, number_bins=4095, ceiling=0.0008):    
    # https://www.cl.cam.ac.uk/teaching/1718/AdvGraph/06_HDR_and_tone_mapping.pdf
    image_histogram, bins = np.histogram(list(filter(lambda v: v < 4090, image.flatten())), number_bins, density=True)
       
    i = 0
    while max(image_histogram) > ceiling and i < 25:
        to_distribute = sum(map(lambda v: v - ceiling, filter(lambda k: k > ceiling, image_histogram)))
        image_histogram = np.fmin(image_histogram, ceiling)
        image_histogram = image_histogram + to_distribute / number_bins
        i+=1
    
    cdf = image_histogram.cumsum() # cumulative distribution function
    cdf = 255 * cdf / cdf[-1] # normalize
    # use linear interpolation of cdf to find new pixel values
    image_equalized = np.interp(image.flatten(), bins[:-1], cdf)

    return image_equalized.reshape(image.shape), cdf

eq_img, cdf = image_histogram_equalization_ceiling(raw_img)

plt.figure()
plt.imshow(255-eq_img, cmap="Greys")
#plt.savefig("flash.png")
plt.show()

plt.figure()
plt.plot(cdf)
plt.show()
```
